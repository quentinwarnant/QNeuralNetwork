```python
import scipy.misc
import matplotlib.pyplot
import imageio.v3
import glob
import numpy
```

```python
rawDataset = []
normalisedDataset = []
```

```python
for image_file_name in glob.glob('Data/processed/2828_my_own_?.png'):
    print("loading ...", image_file_name)
    label = int(image_file_name[-5:-4])
    #load image data from png into array
    img_array = imageio.v3.imread(image_file_name, mode='F')
    #print(img_array)
    #reshape for saving
    img_raw_data = 255.0 - img_array.reshape(784)
    #prefix the label to the start of the raw data
    raw_data = numpy.zeros((785))
    raw_data[0] = label
    raw_data[1:] = img_raw_data
    #reshape from 28x28 to list of 784 values, invert values
    img_data = img_raw_data
    #then scale the data to a range of 0.1 to 1.0
    img_data = (img_data / 255.0 * 0.99) + 0.01
    print(numpy.min(img_data))
    print(numpy.max(img_data))
    #append label and image data to test data set
    record = numpy.append(label, img_data)
    print(record)
    rawDataset.append(raw_data)
    normalisedDataset.append(record)
    pass
```

```python
dataIdx = 2
matplotlib.pyplot.imshow(normalisedDataset[dataIdx][1:].reshape(28,28), cmap='Greys', interpolation='None')
```

```python
numpy.savetxt("Data/q_test.csv", rawDataset, fmt='%i', delimiter=",")
```
